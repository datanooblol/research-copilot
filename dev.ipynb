```python
import arxiv
```

```python
query = "RAG (Retrieval-Augmented Generation) AND (LLM OR Large Language Model) AND (Long Context)"
search = arxiv.Search(
    query=query,
    max_results=5,
    sort_by=arxiv.SortCriterion.Relevance,
    sort_order=arxiv.SortOrder.Descending,
)

client = arxiv.Client()
for result in client.results(search):
    if True:
    # if result.published.year in [2024, 2025]:    
        print(f"Title: {result.title}")
        print(f"Authors: {', '.join(a.name for a in result.authors)}")
        print(f"Published: {result.published}")
        print(f"URL: {result.entry_id}")
        print(f"DOI: {result.doi}")  # Add this line    
        print("-" * 80)
```
